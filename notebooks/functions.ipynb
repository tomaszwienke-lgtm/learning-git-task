# Tworzenie funkcji w Pythonie

Funkcje w Pythonie definiujemy za pomocą słowa kluczowego **`def`**, po którym następuje nazwa funkcji, nawiasy `()` i dwukropek `:`. W kolejnych wierszach (z wcięciem) umieszczamy ciało funkcji, czyli kod, który ma zostać wykonany.

Przykład: funkcja dodająca dwie liczby.

```python
def dodaj(a, b):
    """
    Funkcja zwraca sumę dwóch liczb.
    """
    wynik = a + b
    return wynik
```

# Wywoływanie funkcji

Aby użyć (wywołać) funkcję, piszemy jej nazwę, a w nawiasach okrągłych podajemy argumenty, których oczekuje.

Dla naszej funkcji `dodaj` wywołanie wygląda tak:

```python
# Wywołanie funkcji z argumentami 5 i 3
wynik = dodaj(5, 3)
print("Wynik dodawania:", wynik)

# Możemy też wywołać bezpośrednio w print
print("15 + 7 =", dodaj(15, 7))
```
